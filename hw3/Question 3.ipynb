```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
N = 60
X = np.linspace(-2, 4, N)
Y = np.linspace(-2, 4, N)
X, Y = np.meshgrid(X, Y)

mu = np.array([1,1])
Sigma = np.array([[1,0],[0,2]])
A = multivariate_normal(mu,Sigma)

pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y


Z = A.pdf(pos)

plt.contour(X, Y, Z)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
np.random.seed(1234)
X1 = np.random.normal(3,3,100)
temp = np.random.normal(4,2,100)
X2 = .5*X1 + temp
X1 = np.array(X1).T
X2 = np.array(X2).T
```

```python
mean = [np.mean(X1),np.mean(X2)]
mean
```

```python
tempX1 = np.asmatrix(X1)
tempX2 = np.asmatrix(X2)
matrix = np.append(tempX1,tempX2, axis = 0)
covariance = np.cov(matrix)
covariance
```

```python
eigenvalues, eigenvectors = np.linalg.eig(covariance)
print(eigenvalues)
print(eigenvectors)
```

```python
plt.axis([-15,15,-15,15])
plt.arrow(mean[0],mean[1],eigenvectors[:,0][0]*eigenvalues[0],eigenvectors[:,0][1]*eigenvalues[0],width=.2)
plt.arrow(mean[0],mean[1],eigenvectors[:,1][0]*eigenvalues[1],eigenvectors[:,1][1]*eigenvalues[1],width=.2)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('X1 and X2 plot with eigenvectors')
plt.scatter(X1,X2)
```

```python
U = eigenvectors
zeroedX1 = X1-mean[0]
zeroedX2 = X2-mean[1]
points = np.append(np.asmatrix(zeroedX1),np.asmatrix(zeroedX2),axis=0)
points = np.dot(U.T,points)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Zero and Rotated X1 and X2 plot')
plt.axis([-15,15,-15,15])
plt.scatter(np.array(points[0,:]),np.array(points[1,:]))
```
